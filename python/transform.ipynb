```python
#dependecies
import pandas as pd
```

```python
#importing csv file and creating dataframes
file = 'Resources/kossuth_liquor_sales.csv'
file_2 = 'Resources/kossuth_covid.csv'
liquor_df = pd.read_csv(file, encoding="ISO-8859-1")
kossuth_df = pd.read_csv(file_2, encoding="ISO-8859-1")
```

```python
liquor_df
```

```python
# dropping any rows missing any values
liquor_df = liquor_df.dropna(how='any')
liquor_df
```

```python
# reset the index
clean_liquor_df = liquor_df.reset_index(drop = True)
clean_liquor_df
```

```python
# created the store table
store_df = clean_liquor_df[["storenumber", "storename","address","city", "zipcode", "county"]]
store_df = store_df.drop_duplicates(subset=["storenumber"])
store_df = store_df.reset_index(drop=True)
store_df
```

```python
# created the alcohol table
alcohol_df = clean_liquor_df[["itemnumber", "itemdescription"]]
alcohol_df = alcohol_df.drop_duplicates(subset=["itemnumber"])
alcohol_df = alcohol_df.reset_index(drop=True)
alcohol_df["itemnumber"].value_counts()
```

```python
# sales table
sales_df = clean_liquor_df[["itemnumber","statebottleretail","sale", "invoice", "storenumber","date", "pack", "bottlevolume"]]
sales_df["itemnumber"].value_counts()
```

```python
# kossuth county covid table
kossuth_df = kossuth_df[["date","confirmed","deaths"]]
kossuth_df
```

```python
# export dataframes into excel 
clean_liquor_df.to_excel("clean_kossuth_liquor.xlsx", index=False)
store_df.to_excel("store_tble.xlsx", index=False)
alcohol_df.to_excel("liquor.xlsx", index=False)
sales_df.to_excel("sales.xlsx", index=False)
kossuth_df.to_excel("kossuth_covid.xlsx", index=False)
```
